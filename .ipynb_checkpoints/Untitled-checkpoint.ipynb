### First code simple Markov chain

```python
offset=20.0
stepSize=2.0
lowerBound=0.0
upperBound=100.0

bmc=[]
val=offset
for i in range(0,15000):
    # Add next value as dependent on current state
    val = val + ((random.random()-.5) * stepSize)
    # Need to sample from some distribution so it is more likely to stay near 20
    # Check bounds 
    val = min([max([val,lowerBound]),upperBound])
    bmc.append(val)
```

```python
fig=plt.figure(figsize=[15,5])
plt.plot(bmc)
```

### Add 'normal bounds' - if value is outside normal bounds, it has 30% chance of staying 

```python
offset=20.0
stepSize=2.0
lowerBound=0.0
upperBound=100.0
lowerIBound= 10.0
upperIBound=35.0


bmc=[]
val=offset
for i in range(0,15000):
    # Add next value as dependent on current state
    if val <= lowerIBound:
        val = val + (random.uniform(-1.0/3.0,0.5) * stepSize)
    elif val >= upperIBound:
        val = val + (random.uniform(-0.5,1.0/3.0) * stepSize)
    else:
        val = val + (random.uniform(-0.5,.5) * stepSize)
    # Check bounds 
    val = min([max([val,lowerBound]),upperBound])
    bmc.append(val)
```

```python
fig=plt.figure(figsize=[15,5])
plt.plot(bmc)
```

```python
offset=20.0
stepSize=2.0
maxStepSize=5.0
lowerBound=0.0
upperBound=60.0

# Find function for determining the ranges
sigma = stepSize*7.5
pdfvals=[]
for i in range(int(lowerBound),int(upperBound+1)):
    pdfvals.append(np.exp(-(i-offset)**2 / (2 * sigma**2)))

# Construct Bounded Markov Chain
bmc=[]
val=offset
for i in range(0,86400):
    # Add next value as dependent on current state
    val = val + (random.uniform(-0.5,.5)*maxStepSize*(1.01-pdfvals[int(np.round(val))]))
    # Check bounds 
    val = min([max([val,lowerBound]),upperBound])
    bmc.append(val)

fig=plt.figure(figsize=[15,5])
plt.ylim([0,60])
plt.plot(bmc)
```

```python
class MarkovChain(object):
    def __init__(self, offset, stepSize):
        self.offset = offset
        self.stepSize = stepSize
        self.nextVal = self.offset
    
    def __iter__(self):
        return self
    
    def next(self):
        ret = self.nextVal
        self.nextVal = self.nextVal + (random()-.5) * self.stepSize
    
        return ret
```

```python
class BoundedMarkovChain(object):
    def __init__(self, offset, stepSize, lowerBound, upperBound):
        self.offset = offset
        self.stepSize = stepSize
        self.lowerBound = lowerBound
        self.upperBound = upperBound
        self.begin = True
    
    def bound(self, val):
        if val <= lowerBound:
            ret = lowerBound
        elif self.offset > upperBound:
            ret = upperBound
        else:
            ret = val
        return ret

    def __iter__(self):
        return self
    
    def next(self):
        if self.begin:
            self.currentVal = self.bound(self.offset)
            self.begin = False
        else:
            self.currentVal = self.bound(
                    self.currentVal + (random()-.5) * self.stepSize)
    
        return self.currentVal


    def __init__(self, offset, stepSize, lowerBound, upperBound):
        self.offset = offset
        self.stepSize = stepSize
        self.lowerBound = lowerBound
        self.upperBound = upperBound
        self.mCh = MarkovChain(offset, stepSize)
    
    def __iter__(self):
        return self
    
    def next(self):
        self.nextVal = self.mCh.next()
        if self.nextVal <= self.lowerBound:
            ret = self.lowerBound
        elif self.nextVal > self.upperBound:
            ret = self.upperBound
        else:
            ret = self.nextVal
    
        return ret
```
